```python
# Como la salida del código anterior es Response [200], 
# el servidor da permiso para recoger datos de la página

# Obtenemos el texto del HTML
html_text = requests.get("https://www.komparing.com/es/gasolineras/").text
```

```python
from bs4 import BeautifulSoup


# Parseamos el HTML para convertirlo a un formato más legible
soup = BeautifulSoup(html_text, 'lxml')
```

```python
tabla = soup.find_all('table', id = "tablaGasolineras")
print(tabla)
```

```python
type(tabla)
```

```python
# Obtenemos los títulos de todas las columnas con <td>
headers = []
for t in tabla:
    info_cabecera = t.find_all('td')

# Extraemos el texto en sí
for i in info_cabecera:
    title = i.text
    headers.append(title)

# No sé como excluir de la petición la 'subcabecera'
headers = headers[0:4]
print(headers)
```
